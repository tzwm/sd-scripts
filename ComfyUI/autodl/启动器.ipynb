- ComfyUI 中自动下载的模型路径：`/root/autodl-tmp/ComfyUI/models`
- 同区换实例可以共享使用的模型路径（需要自行创建）：`/root/autodl-fs/models`
- 生成文件路径：`/root/autodl-tmp/ComfyUI/output`
- 基础环境：PyTorch 2.5.1, Python 3.11(ubuntu22.04), Cuda 12.4
